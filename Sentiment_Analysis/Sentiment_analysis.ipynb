# Building a sentiment classifier

## Build word count vectors

```python
products['word_count'] = turicreate.text_analytics.count_words(products['review'])
```

```python
products
```

# Define what is positive and negative sentiment

```python
products['rating'].show()
```

```python
#ignore all 3*  reviews
products = products[products['rating']!= 3]
```

```python
#positive sentiment = 4-star or 5-star reviews
products['sentiment'] = products['rating'] >= 4
```

```python
products
```

```python
products['sentiment'].show()
```

# Train our sentiment classifier

```python
train_data,test_data = products.random_split(.8,seed=0)
```

```python
sentiment_model = turicreate.logistic_classifier.create(train_data,target='sentiment', features=['word_count'], validation_set=test_data)
```

# Apply the sentiment classifier to better understand the Giraffe reviews

```python
products['predicted_sentiment'] = sentiment_model.predict(products, output_type = 'probability')
```

```python
sentiment_model.evaluate(test_data)
```

# Use subset of words to train the model on

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
def count_words(dictionary,word):
    if word in dictionary:
        return dictionary[word]
    else:
        return 0

for word in selected_words:
    products[word] = products['word_count'].apply(lambda x: count_words(x,word), dtype=int, skip_na=True)
```

```python
products.head()
```

```python
word_counts = products.select_columns(selected_words)
```

```python
word_counts = {word: sum(products[word]) for word in selected_words}
```

```python
word_counts
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
selected_words_model = turicreate.logistic_classifier.create(train_data,target='sentiment', features=selected_words, validation_set=test_data)
```

```python
#get coefficients of the words:
coeff = selected_words_model.coefficients.sort('value')
coeff.print_rows(12)
```

```python
selected_words_model_accuracy = selected_words_model.evaluate(test_data)
all_words_model_accuracy = sentiment_model.evaluate(test_data)
```
